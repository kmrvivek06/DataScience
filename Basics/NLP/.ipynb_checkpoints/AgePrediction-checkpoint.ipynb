```python
from nltk.tokenize.casual import EMOTICON_RE
```

```python
def age_features(message):
    emoticon = EMOTICON_RE.findall(' '.join(message))
    length = len(message)
    if not emoticon:
        emoticon = 0
    else:
        emoticon = 1
    return{'Emoticon':emoticon, 'length':length}
```

```python
from nltk.corpus import nps_chat
```

```python
chatroom20 = nps_chat.posts('10-19-20s_706posts.xml')
chatroom30 = nps_chat.posts('10-19-30s_705posts.xml')
chatroom40 = nps_chat.posts('10-19-40s_686posts.xml')
```

```python
chatroom40[54]
```

```python
labeled_chat = ([(post,'20s')for post in chatroom20]+
                [(post,'30s')for post in chatroom30]+
                [(post,'40s')for post in chatroom40])
```

```python
labeled_chat[706]
```

```python
import random
```

```python
random.shuffle(labeled_chat)
```

```python
featureset = [(age_features(n), age) for (n, age) in labeled_chat]
```

```python
train_set, test_set = featureset[500:],featureset[:500]
```

```python
train_set
```

```python
test_set
```

```python
import nltk
```

```python
classifier = nltk.NaiveBayesClassifier.train(train_set)
```

```python
print(nltk.classify.accuracy(classifier,test_set))
```
