```python
def gender_feature(word):
    return{'last_letter': word[-1]}
```

```python
gender_feature('Vivek')
```

```python
from nltk.corpus import names
```

```python
labeled_names = ([(name,'male') for name in names.words('male.txt')]+
                 [(name,'female') for name in names.words('female.txt')])
labeled_names
```

```python
import random
random.shuffle(labeled_names)
```

```python
labeled_names
```

```python
feature_set = [(gender_feature(n), gender) for (n, gender) in labeled_names] 
```

```python
feature_set
```

```python
train_set,test_set=feature_set[500:],feature_set[:500]
```

```python
train_set
```

```python
import nltk
```

```python
classifier=nltk.NaiveBayesClassifier.train(train_set)
```

```python
classifier.classify(gender_feature('Vivek'))
```

```python
classifier.classify(gender_feature('Harsh'))
```

```python
classifier.classify(gender_feature('Ajeet'))
```

```python
print(nltk.classify.accuracy(classifier,test_set))
```
