```python
import tensorflow as tf
from tensorflow.keras.layers import Input, SimpleRNN, Dense, Flatten
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam, SGD
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# make the data
series = np.sin(0.1 * np.arange(200)) + np.random.randn(200)*0.1 #with noise (second part)
# This setup performs good without the noise but fails with noise
#plot
plt.plot(series)
plt.show()
```

```python
############################################
# N = No. of samples
# T = Sequence or time series length
# D = No. of features
# M = No. of hidden units
# K = No. of output units
############################################
```

```python
# build the data set
T = 10
X = []
Y = []
for t in range(len(series) - T):
    x = series[t:t+T]
    y = series[t+T]
    X.append(x)
    Y.append(y)
    
X = np.array(X).reshape(-1, T, 1)   # Data should be 3 dimention N X T X D (No. of samples, Timse series, Features)
Y = np.array(Y)

N = len(X)
print("X.shape", X.shape, "Y.shape", Y.shape)
```

```python
# build RNN model
i = Input(shape=(T,1))
#x = SimpleRNN(5)(i)
x = SimpleRNN(5, activation='relu')(i)
x = Dense(1)(x)
model = Model(i, x)
model.compile(loss='mse', optimizer=Adam(lr=0.1))
```

```python
r = model.fit(X[:-N//2], Y[:-N//2], validation_data=(X[-N//2:], Y[-N//2:]), epochs=80)
```

```python
# Plot loss per iteration
plt.plot(r.history['loss'], label='loss')
plt.plot(r.history['val_loss'], label='val_loss')
plt.legend()
plt.show()
```

```python
# "wrong" forecast using true targets
validation_target = Y[-N//2:]
validation_predictions = []

# index of first validation
i = -N//2

while(len(validation_predictions) < len(validation_target)):
    P = model.predict(X[i].reshape(1,-1,1))[0,0]
    i = i+1
    validation_predictions.append(P)
```

```python
# Plot the predictions
plt.plot(validation_target, label='validation_target')
plt.plot(validation_predictions, label='validation_predictions')
plt.legend()
plt.show()
```

```python
# "Correct" forecast using self predicted
validation_target = Y[-N//2:]
validation_predictions = []

# index of first validation
last_x = X[-N//2]

while(len(validation_predictions) < len(validation_target)):
    P = model.predict(last_x.reshape(1,-1,1))[0,0]
    validation_predictions.append(P)
    last_x = np.roll(last_x, -1)
    last_x[-1] = P
```

```python
# Plot the predictions
plt.plot(validation_target, label='validation_target')
plt.plot(validation_predictions, label='validation_predictions')
plt.legend()
plt.show()
```
